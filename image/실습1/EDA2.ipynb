```python
# 라벨 별로 한개씩 이미 확인
# 티셔츠 한개 찾아서, 이미지 확인
img=train_df[train_df['label']==0].iloc[0,:]
img=img.drop(['index','label'])
print(img.shape)
img = np.array(img,dtype='float32')
plt.imshow(img.reshape(28,28),cmap='gray')
```

```python
# 0~10번 까지 이미지 그려보기

plt.figure(figsize=(20,50))
for i in range(10):
    img=train_df[train_df['label']==i].iloc[0,:]
    print(train_df[train_df['label']==i].shape)
    img=img.drop(['index','label'])
    img = np.array(img,dtype='float32')
    plt.subplot(1,10,i+1) # row, column, index
    plt.imshow( img.reshape(28,28) , cmap='gray')
```
